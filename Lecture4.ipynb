# Physical models: 

So we'll just use dimensional analysis for studying drag because who wants to use navier-stokes??? 

Relevant variables: 
Size of object: r [L]
Velocity of object: v [L/T]
Fluid density: $\rho$: [M/L$^3$]
Kinematic viscosity $\nu$: [L$^3$/T]
    (there's also a dynamic viscosity you might see, but not important)

There's something called Buckingham-Pi theorem that'll be useful.

Want to find a dimensionless quantity:
$$\frac{vr}{\nu}$$
Will work, and ends up being the Reynolds number! Whaddya know. 
$$=\frac{\text{inertial forces}}{\text{viscous forces}} $$

Now we want to find the characteristic time scale, which is vital to know for simulation (time step should be << char time)

$$ \tau_v \equiv \frac{r}{v_0} $$
$$\tau_\nu \equiv \frac{r}{V} = \frac{r^2}{\nu}$$

Wo if you know things about random walks: 
$$<r^2> \propto \nu \tau_\nu $$
Looks like a characteristic equation for a random walk. So $\nu$ can be interpreted as a diffusion constant for momentum in the fluid. 

$$Re \equiv \frac{\tau_\nu}{tau_v}$$

In the limit $Re<<1$: [called limit of Laminar flow] \\
There's plenty of time for the fluid to equilibriate

 In the limit $Re>>1$: [called the limit of turbulent flow]\\
 There's no time for information to travel from one part to another 
 Viscosity is no longer an important parameter 

( I think $Re<<1$?)
 $$\Delta x  = v_0 \Delta t + \frac{1}{2} \frac{f}{M} <\Delta t>^2$$
 $$<\Delta x > = <v_0>\Delta t + \frac{1}{2} \frac{f}{M} <\Delta t>^2$$
 $$= \frac{1}{2} \frac{f}{M} <\Delta t>^2$$
 $$v_{} = \frac{<\Delta x>}{\Delta t} = \frac{f}{\zeta} $$
 Where $\zeta = 2m/\Delta t$
 $$F_d = -f= \rho v \,\,\,(?)$$

# Object oriented programming 

People used to think it was the greatest thing ever, but now there's some controversy over how good it is based on maintainability

Procedural programming:
- Is NOT object oriented 
- operates on data through a series of functions  
- fundamental units are functions 

Object oriented programming: 
- A program is structured with different objects, including: 
    Functions: maps some input to some output 
- Fundamental units are objects 
    (abstract representation of a thing)
    It has attributes can be data or properties, called fields 
    An attribute can also be functions (called methods in python)
- Python is a CLASS based language 
    A class is a blueprint for specific object
- An object is an INSTANCE of a class 
- Objects in python also has a property called INHERITANCE 
    So you can create a subclass that inherits all properties from parent class
